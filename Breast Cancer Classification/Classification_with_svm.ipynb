```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
df_cancer=pd.DataFrame(np.c_[cancer['data'],cancer['target']],columns=np.append(cancer['feature_names'],['target_names']))
df_cancer.head(10)
```

```python
df_cancer.shape
```

Corelation between Features

```python
plt.figure(figsize=(25,15))
sb.heatmap(df_cancer.corr(),annot=True)
```

```python
df_cancer.columns
```

```python
X=df_cancer.drop(columns='target_names')
X.head(10)
```

```python
y=df_cancer['target_names']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=102)
X_train.head(10)
```

NORMALISING DATASET

```python
scale_of_X_train =(X_train-X_train.min())/(X_train.max()-X_train.min())
scale_of_X_train.head(10)
```

```python
scale_of_X_test =(X_test-X_test.min())/(X_test.max()-X_test.min())
scale_of_X_test.head(10)
```

```python
from sklearn.svm  import SVC
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
svc=SVC()
svc.fit(scale_of_X_train,y_train)
y_predict=svc.predict(scale_of_X_test)
```

CONFUSION MATRIX

```python
matrix=np.array(confusion_matrix(y_test,y_predict))
conf_matrix=pd.DataFrame(matrix,columns=['Cancer Patients','Healthy Persons'],index=['predicted cancer','predicted health'])
conf_matrix
```

```python
y_test.value_counts()
```

```python
sb.heatmap(conf_matrix,annot=True,fmt=".0f")
```

CLASSIFICATION REPORT

```python
print(classification_report(y_test,y_predict))
```

ACCURACY SCORES

```python
print(accuracy_score(y_test,y_predict))
```
