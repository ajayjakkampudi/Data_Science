CLEANING THE TEXT

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus=[]
c=[]
for i in range(0,1000):
    review=re.sub('[^a-zA-Z]',' ',df['Review'][i])
    review=review.lower()
    review=review.split()
    #review=' '.join(review)
    
    
    
    all_stopwords = stopwords.words('english')
    #all_stopwords.remove("won't")
    removing_words=['ain',
 'aren',
 "aren't",
 'couldn',
 "couldn't",
 'didn',
 "didn't",
 'doesn',
 "doesn't",
 'hadn',
 "hadn't",
 'hasn',
 "hasn't",
 'haven',
 "haven't",
 'isn',
 "isn't",
 'ma',
 'mightn',
 "mightn't",
 'mustn',
 "mustn't",
 'needn',
 "needn't",
 'shan',
 "shan't",
 'shouldn',
 "shouldn't",
 'wasn',
 "wasn't",
 'weren',
 "weren't",
 'won',
 "won't",
 'wouldn',
 "wouldn't", 'don',
 "don't",'too',
 'very','no','same',
 'so',
 'than',
 'too',
 'very',
 'nor',
 'not','y'
 'only']

           
    
    for i in range(0,len(removing_words)):
        if removing_words[i] in all_stopwords:
            all_stopwords.remove(removing_words[i])
    ps=PorterStemmer()
    review = [ps.stem(word) for word in review if not word in set(all_stopwords)]

    final_review = ' '.join(review)
    
    c.append(review)
    corpus.append(final_review)
    





```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1500)
X = cv.fit_transform(corpus).toarray()
y = df.iloc[:, -1].values
```

```python
len(X[0])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 100)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(C=1.623776739188721)
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
#print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
